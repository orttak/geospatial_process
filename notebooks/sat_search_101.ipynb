```python
bbox = [24, 34, 50, 44] # (min lon, min lat, max lon, max lat)
dates = '2020-07-01/2020-07-12'

URL='https://earth-search.aws.element84.com/v0'
results = satsearch.Search.search(url=URL,
                                  collections=['sentinel-s2-l2a-cogs'],
                                  datetime=dates,
                                  bbox=bbox,
                                  query={'eo:cloud_cover': {'lt': 10},'sentinel:utm_zone':{'gte':35,'lt':39}},
                                  #sortby=["eo:cloud_cover"]
                                  #asc
                                  #sortby=[ {"field": "eo:cloud_cover","direction": "desc" }],
                                  #sort=[{'cloud_cover':'desc'},{'datetime':'asc'}]    
                                  #sortby=['datetime']
                                  )
```

```python
items = results.items()
```

```python
sentinel_list=['35TME','35TNE','35SMD','35SND']
```

```python
for item in items:
    id=item.id.split('_')[1]
    if id in sentinel_list:
        print(item.assets['B04']['href'])
```

```python
print(items.summary(['date', 'id', 'eo:cloud_cover']))
```

```python
items.filter('sentinel:grid_square',['YB','WB'])
```

```python
items.filter('sentinel:utm_zone',[36,])
items.filter('sentinel:latitude_band',['M'])
items.filter('sentinel:grid_square', ['YB'])
selected_item=sorted(items.properties('eo:cloud_cover'))[0]
items.filter('eo:cloud_cover', [selected_item])
print(items.summary(['date', 'id', 'eo:cloud_cover']))
items.__getitem__(0)
```

```python
# Use this blok when user wants more then 1 image for same Sentinel tile
#user demand
numberOfImages=3
items.filter('sentinel:utm_zone',[36,])
items.filter('sentinel:latitude_band',['M'])
items.filter('sentinel:grid_square', ['YB'])
selected_items=sorted(items.properties('eo:cloud_cover'))
items.filter('eo:cloud_cover',selected_items[0:numberOfImages] )
print(items.summary(['date', 'id', 'eo:cloud_cover']))
```
